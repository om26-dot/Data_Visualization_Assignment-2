2)Develop a code to plot bivariate distribution considering a suitable
data set available on the open source

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load an open-source dataset
data = sns.load_dataset('iris')  # Iris dataset

# Plot bivariate distribution (e.g., Sepal Length vs Sepal Width)
sns.kdeplot(data=data, x="sepal_length", y="sepal_width", cmap="Blues", fill=True)

# Add labels and title
plt.title('Bivariate Distribution: Sepal Length vs Sepal Width')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')

# Show the plot
plt.show()
```
